```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
print(df[df['TotalCharges'].isnull()][['tenure', 'MonthlyCharges']])
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0.0)
```

```python
df['Churn_Numeric'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['tenure'], kde=True, ax=axes[0], color='slategray').set_title('Tenure Distribution')
sns.histplot(df['MonthlyCharges'], kde=True, ax=axes[1], color='crimson').set_title('Monthly Charges Distribution')
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(x='Contract', y='tenure', hue='Churn', data=df, palette='Set2')
plt.title('Tenure Distributions across Contract Types Split by Churn Outcome')
plt.ylabel('Tenure (Months)')
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn_Numeric']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Continuous Metric Feature Correlation Blueprint')
```
